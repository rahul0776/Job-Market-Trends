## MERGING DATASETS

```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv('./datasets/kaggle-survey-2019/multiple_choice_responses.csv')  # multiple_choice_responses dataset
df2 = pd.read_csv('./datasets/US_graduates/US_graduates.csv')               # US_graduates dataset
df3 = pd.read_csv('./datasets/Job_Postings/postings_half.csv')                   # postings dataset
df4 = pd.read_csv('./datasets/AI_Job_Market/ai_job_market_insights.csv')              # ai insights dataset

# Print column names to debug
print("df1 (multiple_choice_responses) columns:", df1.columns)
print("df2 (US_graduates) columns:", df2.columns)
print("df3 (postings) columns:", df3.columns)
print("df4 (ai_insights) columns:", df4.columns)
```

```python
# Rectifying header of dataframe 1 
df1.columns = df1.iloc[0]
# Drop the second row (as it is now the header)
df1 = df1.drop(0)

df1.rename(columns={
    "In which country do you currently reside?": "Location",
    "Select the title most similar to your current role (or most recent title if retired): - Selected Choice": "Job_Title",
}, inplace=True)
```

```python
# Merge the datasets on common columns like 'Job_Title' and 'Location'
merged_df = df1.merge(df3, on=['Job_Title', 'Location'], how='outer')\
                .merge(df4, on=['Job_Title', 'Location'], how='outer')\
                .merge(df2, left_on='Job_Title', right_on='Education.Major', how='outer')

# Display the first few rows of the merged dataset
print(merged_df.head())

# Save the merged dataset to a CSV file
merged_df.to_csv('merged_dataset.csv', index=False)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
```

```python
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

```python
def detect_and_ordinal_encode(df):
    # Create a copy of the original dataframe
    df_encoded = df.copy()
    
    # Initialize a dictionary to store label encoders for each column
    label_encoders = {}

    # Loop through columns
    for col in df_encoded.columns:
        # Check if column is categorical (object type)
        if df_encoded[col].dtype == 'object':
            # Apply Label Encoding
            le = LabelEncoder()
            df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
            
            # Store the LabelEncoder for future use (e.g., for decoding or applying to test data)
            label_encoders[col] = le

    return df_encoded, label_encoders

encoded_df , labels_encoders = detect_and_ordinal_encode(merged_df)
```

```python
knn_imputer = KNNImputer(n_neighbors=5)

# Perform imputation and convert the result back to a DataFrame
df_imputed = pd.DataFrame(knn_imputer.fit_transform(encoded_df), columns=encoded_df.columns)

# Ensure the encoded categorical columns are of integer type after imputation (since KNN Imputer returns floats)
df_imputed = df_imputed.apply(lambda x: x.astype(int) if x.name in labels_encoders else x)
```
